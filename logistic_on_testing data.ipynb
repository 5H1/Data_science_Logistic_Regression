```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Loading the training data set
training_data = pd.read_csv('Titanic_train.csv')
```

```python
##handle the missing values
training_data['Age'].fillna(training_data['Age'].median(), inplace=True)
training_data['Embarked'].fillna(training_data['Embarked'].mode()[0], inplace=True)

# we are going to Drop the 'Cabin' column due to a large number of missing values
training_data.drop(columns=['Cabin'], inplace=True)

# Encoding the categorical variables
training_data = pd.get_dummies(training_data, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
# Define the features for X and the target variable here is "Survived"
X_train = training_data.drop(columns=[ 'Name', 'Ticket', 'Survived'])
y_train = training_data['Survived']
```

```python
# Training the logistic regression model
log_regr = LogisticRegression(max_iter=1000)
log_regr.fit(X_train, y_train)
```

```python
# Loading the test data set
test_data = pd.read_csv('Titanic_test.csv')

# Handling the missing values in the dataset
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)

# Encodig for categorical variables
test_data = pd.get_dummies(test_data, columns=['Sex', 'Embarked'], drop_first=True)

# Ensurign that the test dataset has the same features as present in the training dataset
# Addding the missing dummy variables if necessary
missing_colmns = set(X_train.columns) - set(test_data.columns)
for col in missing_colmns:
    test_data[col] = 0

# Reordering the columns as same of present in training data set
test_data = test_data[X_train.columns]
```

```python
# Making the predictions on the test dataset
test_predictions = log_regr.predict(test_data)

# Creating a DataFrame with these predictions
submission_test = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': test_predictions})

# Saving the predictions to a CSV file
submission_test.to_csv('test_submission.csv', index=False)
```
